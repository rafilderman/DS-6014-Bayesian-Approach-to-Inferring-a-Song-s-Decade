```python
# All Out 80s
playlist_id = '37i9dQZF1DX4UtSsGT1Sbe'
url = 'https://api.spotify.com/v1/playlists/37i9dQZF1DX4UtSsGT1Sbe/tracks'
spotify_request = requests.get(url, params)
spotify_request
```

```python
spotify_json = json.loads(spotify_request.text)
playlist_json = spotify_json['items']
songs_df = pd.DataFrame([s['track']['id']] for s in playlist_json)
trackids = songs_df[0].to_list()
tracklistToStr = ','.join(map(str, trackids)) 
```
